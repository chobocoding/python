# heapq 모듈 사용법

## 힙 자료구조
heapq 모듈은 이진 트리 기반의 **최소 힙** 자료구조를 제공한다.

image.png

```python
import heapq

# 힙 생성
heap = []

# 힙에 원소 추가 heappush(리스트, 추가할 원소)
heapq.heappush(heap, 4)
heapq.heappush(heap, 1)
heapq.heappush(heap, 5)
heapq.heappush(heap, 8)
heapq.heappush(heap, 0)
print(heap)

# heappush() 함수는 O(logN)의 시간 복잡도
```

```python
# 기존에 리스트에 값이 있을 경우 최소힙 정렬이 안됨
ls = [1, 5, 7, 8, 9, 10, 5, 6]
heapq.heappush(ls, -5)
print('정렬 X :', ls)

# 기존 리시트를 힙으로 변환하려면 heapify()를 이용해야 함
ls = [1, 5, 7, 8, 9, 10, 5, 6]
heapq.heapify(ls)
heapq.heappush(ls, -5)
print('정렬 O :',ls)

# heapify() 함수의 성능은 인자로 넘기는 리스트의 원소수에 비례합니다. 즉 O(N)의 시간 복잡도
```

```python
# 힙에서 원소 삭제 heapq 모듈의 heappop() 함수를 이용
print(heapq.heappop(heap))
print(heap)
# heappop() 함수도 역시 O(logN)의 시간 복잡도
```

```python
# 인덱싱도 가능함
print(heap[0], heap[-1])
```
