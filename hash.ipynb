```python
# Dictionary 사용법
# 1. dictionary = dict()
# 2. dictionary = {}
dict1 = dict()
dict2 = {}
print(type(dict1), type(dict2))
```

```python
# key - value 쌍을 갖는 dictionary 선언
name = {'성' : '조',
       '이름' : '현준'}
name
```

```python
# 원소 가져오는 방법
# 1. []
# 2. get 메소드

print(name['성'])
```

```python
# 없는 키 값을 입력할 경우
print(name['??'])
```

```python
print(name.get('성'))
```

```python
# 없는 키 값을 입력할 경우
print(name.get('??'))
```

```python
# 없는 키 값일 때 출력할 내용을 넣을 수 있음
print(name.get('??', '해당 키값 없음'))
print(name.get('이름', '해당 키값 없음'))
```

```python
# 업데이트
name['성2'] = '추가'
name['이름2'] = '민섭'
name['성'] = '업데이트'
name
```

```python
# 삭제
# 1. del dict[key]
# 2. pop(key, [default])

del name['성2']
name
```

```python
# 키가 없는 경우 del로 삭제
del name['성2']
name
```

```python
name.pop('성')
print(name)
```

```python
# 키가 없는 경우 pop으로 삭제
name.pop('성', 'default')
```

```python
# Iteration
# 1. key
# 2. Key, value 둘 다

for key in name:
    print(key)

for key, value in name.items():
    print(key, value)
```
